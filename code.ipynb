```python
df_country =pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
df_final =pd.merge(df,df_country, on='Country Code', how='left')
```

```python
df_final.dtypes
```

```python
df_final.Country.value_counts()
```

```python
country_names=df_final.Country.unique()
len(df_final.Country.unique())
```

```python
country_names
```

```python
df_final.Country.value_counts().values
country_values=df_final.Country.value_counts().values
country_values
```

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct="%.2f%%")
```

##### Observation:Zomato maximum records or transaction are from India with 94.39 percentage preceeded by USA and then United Kingdoms

```python
ratings=df_final. groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings
```

Observation
1. When Rating is between 4.5 to 4.9--> Excellent
2. When Rating are between 4.0 to 3.4 >very good
3. when Rating is between 3.5 to 3.9--> good
4. when Rating is between 3.0 to 3.4 > average
5. when Rating is between 2.5 to 2.9-> average
6. when Rating is between 2.0 to 2.4 > Poor

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue="Rating color" ,data=ratings, palette=[ 'silver', 'red', 'orange', 'yellow', 'green', 'green' ])
```
